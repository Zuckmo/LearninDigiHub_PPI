## Sampling Method

https://raw.githubusercontent.com/krishna0604/dataset_udemy/refs/heads/main/ab_data.csv

```python
df_ab = pd.read_csv("https://raw.githubusercontent.com/krishna0604/dataset_udemy/refs/heads/main/ab_data.csv")
df_ab.shape
```

```python
df_ab.head()
```

Simple Random Sampling

Ingin mengetahui dari hasil sampel sebesal 400 data, berapakah yang menerima test old_page dan berapakah yang menerima new_page.

```python
sample_df = df_ab.sample(400)
sample_df["landing_page"].value_counts()
```

Ketika kode diatas di running lebih dari 1 kali, maka hasilnya akan berbeda. Jika hasilnya berbeda maka teknik sampling nya juga akan berbeda setiap kali running, akan menyebabkan kesalahan analisis. Ccara mengatasinya adalah :

```python
# Menfilter old_page dan new_page
old_page_df = df_ab[df_ab["landing_page"] == "old_page"]
new_page_df = df_ab[df_ab["landing_page"] == "new_page"]

# mengaplikasikan sistem random sampling
sample_old_page = old_page_df.sample(200)
sample_new_page = new_page_df.sample(200)

# menyatukan kedua variabel menjadi satu variabel menggunakan .concat()
sample_df = pd.concat([sample_old_page, sample_new_page], axis = 0) # data yang ditambah akan berpengaruh pada baris (axis = 0), kalau axis = 1, maka kolomnya yang bertambah
sample_df["landing_page"].value_counts()
```

Berhasil menetapkan bahwa kodenya jika di run terus jumlah samplenya tetap sama

```python
sample_df["landing_page"].head()
```

Awalnya, data teratas adalah data dengan nomor indeks 77252, namun ketika kode sampling di run, data teratas berubah lagi menjadi nomor indeks 270378. Hal ini akan menyebabkan kesalahan anlisis lagi, maka apa yang dapat dilakukan? kita akan mengimplementasikan random_state

```python
# mengaplikasikan sistem random sampling
sample_old_page = old_page_df.sample(200, random_state = 123)
sample_new_page = new_page_df.sample(200, random_state = 123)

# menyatukan kedua variabel menjadi satu variabel menggunakan .concat()
sample_df = pd.concat([sample_old_page, sample_new_page], axis = 0) # data yang ditambah akan berpengaruh pada baris (axis = 0), kalau axis = 1, maka kolomnya yang bertambah
sample_df["landing_page"].head()
```

Awalnya indeks teratas adalah 57982, ketika di run lagi indeks teratas tetap sama di angka 57982. Maka kita berhasil menerapkan random sampling
