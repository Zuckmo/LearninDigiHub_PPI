# Merge

```python
dict_a = pd.DataFrame({"no": [1, 2],
                         "nama": ["John", "Judy"],
                         "umur": [24, 20]})
dict_b = pd.DataFrame({"no": [1, 3],
                       "nama":["John", "Adi"],
                       "alamat": ["Jakarta", "Bandung"]})
```

Kita pengen menampilkan satu tabel yang terdapat data no, nama, umur, dan alamat

```python
df_merge_1 = dict_a.merge(dict_b,
                          left_on = "nama",
                          right_on = "nama",
                          how = "inner")
# inner join : irisan dari kedua dict
df_merge_1
```

```python
df_merge_1 = dict_a.merge(dict_b,
                          left_on = "nama",
                          right_on = "nama",
                          how = "left")
# inner join : irisan dari kedua dict
df_merge_1
```

```python
df_merge_1 = dict_a.merge(dict_b,
                          left_on = "nama",
                          right_on = "nama",
                          how = "right")
# inner join : irisan dari kedua dict
df_merge_1
```

Kasus 3 : membuat kalkulator penjumlahan dan pengurangan

```python
def kalkulator (metode, a, b):
  metode = metode.lower()
  if metode == 'penjumlahan':
    hasil = a+b
    return hasil #return statement 1
  elif metode == 'pengurangan':
    hasil = a-b
    return hasil #return statement 2
  else :
    hasil = '-'
  print('inputan metode', metode, 'dengan inputan', a, 'dan', b, 'tidak sesuai inputan', hasil)

print(kalkulator ('penjumlahan', 5, 6))
print(kalkulator('penjum', 5,6))

# jika suatu kondisi sudah memenuhi, function berhenti. hasil:
# 11
# inputan metode penjum dengan inputan 5 dan 6 tidak sesuai inputan -
# None
```

```python
# syntax I
# misal ada tabel dengan kolom a [on_going, on_going, on_going] dan kolom b [False, True, True]
# Data awal
tabel = pd.DataFrame({'a': ["uncomplete", "on_going", "on_going"],
                      'b': [False, True, True]})

# Fungsi untuk menentukan nilai kolom 'c'
def aksi(baris):
    if baris['b'] == True:
        return 'selesai'
    else:
        return baris['a']

# Menggunakan apply() untuk menerapkan fungsi ke setiap baris
tabel['c'] = tabel.apply(aksi, axis=1)

print(tabel)
```

```python
import pandas as pd

# Data awal
tabel = pd.DataFrame({'a': ["uncomplete", "on_going", "on_going"],
                      'b': [False, True, True]})

# Ubah nilai kolom 'a' menjadi 'selesai' jika kolom 'b' bernilai True dan 'a' adalah 'on_going'
tabel['c'] = tabel.apply(lambda row: 'selesai' if row['b'] and row['a'] == 'on_going' else row['a'], axis=1)

print(tabel)
```
